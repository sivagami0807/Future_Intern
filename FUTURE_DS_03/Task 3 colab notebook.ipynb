```python
def generate_feedback(rating):
    if rating <= 2:
        return "The course experience was not satisfactory and needs improvement."
    elif rating == 3:
        return "The course experience was average and could be improved."
    else:
        return "The course experience was very good and highly informative."

df['Feedback_Text'] = df['Recommendation'].apply(generate_feedback)
```

```python
df[['Recommendation', 'Feedback_Text']].head(10)
```

```python
from textblob import TextBlob
def textblob_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
df['TextBlob_Sentiment'] = df['Feedback_Text'].apply(textblob_sentiment)
df[['Feedback_Text', 'TextBlob_Sentiment']].head(10)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure()
sns.countplot(x='TextBlob_Sentiment', data=df)
plt.title("TextBlob Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Number of Students")
plt.show()
```
